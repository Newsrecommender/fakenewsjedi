```python
# Load libraries
import pandas 
import statsmodels.api as sm
import pylab as pl
import numpy as np
from sklearn.model_selection import train_test_split

# Load dataset

whole = pandas.DataFrame.from_csv("Facebook_scrape1.csv")



dummies1 = pandas.get_dummies(whole['Post Type'])
dummies2 = pandas.get_dummies(whole['Rating'])
names = ['account_id','reaction_count','comment_count']
clean = whole[names].join(dummies1)
data = clean.join(dummies2)
new = pandas.concat([whole, dummies1,dummies2], axis=1)

data['intercept'] = 1.0
data, testList= train_test_split(data,test_size = .2) 
testList = testList.dropna()
data
```

```python
df1 = data.dropna()
df1
train_cols = (df1.columns[0:2]).append(df1.columns[2:5]).append(df1.columns[10:])

print (train_cols)
logit = sm.Logit(df1['mostly true'], df1[train_cols])

result = logit.fit()
print (result.summary())
```

```python
print (result.conf_int())
```

```python
print (np.exp(result.params))
```

```python
params = result.params
conf = result.conf_int()
conf['OR'] = params
conf.columns = ['2.5%', '97.5%', 'OR']
print (np.exp(conf))
```

```python
testList['mostly'] = result.predict(testList[train_cols])
```
